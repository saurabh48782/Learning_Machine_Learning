```python
import pandas as pd
from sklearn import model_selection
```

```python
# Reading the dataset
data = pd.read_csv('Iris.csv')
data.head()
```

```python
# Creating a new column in the dataset having name 'kfold' and filling all the field with the value -1 
data['kfold']=-1
data.head()
```

```python
# Randomizing the rows inside the datset and resetting the index of the rows
data = data.sample(frac=1).reset_index(drop=True)
data.head()
```

```python
# Initializing the k-fold from the model_selection package in sklearn library
kf = model_selection.KFold(n_splits=5) # splitting the dataset into 5 equal parts
```

```python
# Filling the new kfold column
for fold, (train_,val_) in enumerate(kf.split(X=data)):
    data.loc[val_,'kfold']=fold
    
# Saving the new csv with kfold column
data.to_csv('Iris_Kfold.csv',index=False)
```
