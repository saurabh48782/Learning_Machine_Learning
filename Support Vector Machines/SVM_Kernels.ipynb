```python
from sklearn import svm
```

  ### **sklearn's svm module supports three different types of classifiers which are:**
  1. SVC is the SVM which supports kernels.
  2. NuSVC is used when we want to reduce the number of support vectors.
  3. LinearSVC is implemented using the Pegasaus algorithm.

### Kernel based separations in SVM

sklearn supports the folowing types of kernels, which can be used in many real-life problems.

 - Linear Kernel
 - RBF Kernel
 - Polynomial Kernel
 - Sigmoid Kernel
 
**The default kernel used in SVC is always the RBF (Radial Basis Function) kernel.**

**Kernel Trick** is method of using a linear classifier to solve a non-linear problem. It transforms the linearlly inseparable data into linearlly separable data.


##### Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

#### Dataset Preparation

```python
X,Y = make_circles(n_samples=500,noise=0.02)
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.grid(True)
plt.show()
```

#### Classifying this non-linearlly separable dataset using SVM

```python
# SVC supports kernel version
svc = svm.SVC(kernel="linear")
```

```python
# Using Linear Kernel
svc.fit(X,Y)
```

```python
accuracy = svc.score(X,Y)
print("Accuracy obtined using a linear kernel is: ",accuracy)
```

```python
# Using RBF Kernel
svr = svm.SVC(kernel="rbf")
```

```python
svr.fit(X,Y)
```

```python
accuracy = svr.score(X,Y)
print("Accuracy obtined using a rbf kernel is: ",accuracy)
```

```python
# Using Polynomial Kernel
svp = svm.SVC(kernel='poly')
svp.fit(X,Y)
accuracy = svp.score(X,Y)
print("Accuracy obtined using a polynomial kernel is: ",accuracy)
```

```python
# Making a Custom Kernel for classification
def custom_kernel(x1,x2):
    return np.square(np.dot(x1,x2.T))

svck = svm.SVC(kernel=custom_kernel)
svck.fit(X,Y)
accuracy = svck.score(X,Y)
print("Accuracy obtined using a custom kernel is: ",accuracy)
```
